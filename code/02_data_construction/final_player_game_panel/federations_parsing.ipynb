# FIDE Metadata Parsing

This notebook replaces the old Selenium scraping workflow.

It uses `07_find_player_metadata_from_fide.py`, which can:

- start from Chess.com usernames or from real names already present in a CSV,
- use the Chess.com public API to recover real names and titles when needed,
- match those players against the official FIDE XML export,
- output federation, real name, standard/rapid/blitz ratings, and age.

Before running the script, download the latest official FIDE XML ZIP from:
`https://ratings.fide.com/download_lists.phtml`
and place it at `/tmp/players_list_xml.zip` or update the path below.

```python
from pathlib import Path

import pandas as pd

INPUT_CSV = Path("output/players_in_regression_missing_from_players_final_data.csv")
OUTPUT_CSV = Path("output/player_metadata_from_fide_missing_players.csv")
FIDE_XML_ZIP = Path("/tmp/players_list_xml.zip")
AGE_AS_OF_YEAR = 2026

INPUT_CSV, OUTPUT_CSV, FIDE_XML_ZIP
```

```python
import subprocess

command = [
    "python3",
    "07_find_player_metadata_from_fide.py",
    "--input",
    str(INPUT_CSV),
    "--output",
    str(OUTPUT_CSV),
    "--fide-xml-zip",
    str(FIDE_XML_ZIP),
    "--age-as-of-year",
    str(AGE_AS_OF_YEAR),
]

subprocess.run(command, check=True)
```

```python
result = pd.read_csv(OUTPUT_CSV)
result.head(20)
```

## Notes

- If your input CSV already contains real names, add `--skip-chesscom` to avoid API calls.
- The script auto-detects common input layouts:
  - `white_name` / `black_name` pairings files,
  - `player_name` player lists,
  - `real_name` or `name` real-name lists.
- Output columns include match diagnostics so unresolved or ambiguous players are easy to review.
