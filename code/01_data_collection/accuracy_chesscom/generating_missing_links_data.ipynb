# Adding missed links

```python
scraped_accuracy_1 = pd.read_csv('scrape_outputs_second_iteration/worker_0_checkpoint_40660.csv')
scraped_accuracy_2 = pd.read_csv('scrape_outputs_second_iteration/worker_0_checkpoint_53505_reverse.csv')
```

```python
scraped_accuracy_1 = scraped_accuracy_1[scraped_accuracy_1['ok'] == True][['links', 'accuracy1', 'accuracy2']].reset_index(drop=True).rename(columns={'accuracy1': 'accuracy_white', 'accuracy2': 'accuracy_black'})
scraped_accuracy_2 = scraped_accuracy_2[scraped_accuracy_2['ok'] == True][['links', 'accuracy1', 'accuracy2']].reset_index(drop=True).rename(columns={'accuracy1': 'accuracy_white', 'accuracy2': 'accuracy_black'})
scraped_accuracy = pd.concat([scraped_accuracy_1, scraped_accuracy_2]).reset_index(drop=True)
scraped_accuracy
```

```python
tournaments = tournaments.merge(
    scraped_accuracy[['links', 'accuracy_white', 'accuracy_black']],
    on='links',
    how='left',
    suffixes=('', '_new')
)

tournaments['accuracy_white'] = tournaments['accuracy_white'].fillna(tournaments['accuracy_white_new'])
tournaments['accuracy_black'] = tournaments['accuracy_black'].fillna(tournaments['accuracy_black_new'])

tournaments = tournaments.drop(columns=['accuracy_white_new', 'accuracy_black_new'])
```

```python
tournaments
```

```python
new_links = tournaments[(tournaments['accuracy_white'].isna()) & (tournaments['accuracy_black'].isna())].reset_index(drop=True)['links']
```

```python
new_links.to_csv('new_missed_links_2026_third_iteration.csv')
```
